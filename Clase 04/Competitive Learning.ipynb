```python
import pandas as pd
import numpy as np
```

```python
#PASO 1. CARGAR BASE DE DATOS Y RUTAS
BaseAumentada = pd.read_csv("C:/Users/clau_/OneDrive/Documentos/FCFM/03_Tetramestre/Aprendizaje_Profundo/Clase 04/BaseLD.csv")
BaseAumentada
```

```python
# Inicializar matrices
m, n = BaseAumentada.shape
neuronas = np.zeros((3, n))
promedio = np.zeros((1, n))
minimo = np.zeros((1, n))
maximo = np.zeros((1, n))
pertenencia = np.zeros((m, 1))
```

```python
# Initialize the locations
for i in range(n):
    minimo[0, i] = BaseAumentada.iloc[:, i].min()

for i in range(n):
    maximo[0, i] = BaseAumentada.iloc[:, i].max()

for i in range(n):
    promedio[0, i] = BaseAumentada.iloc[:, i].mean()

neuronas[0, :] = minimo[0, :]
neuronas[1, :] = promedio[0, :]
neuronas[2, :] = maximo[0, :]
```

```python
# Synaptic Potential
distancia = np.zeros((1, 3))
base = BaseAumentada.values
```

```python
# Cluster
cluster = np.zeros((1, 3))

for k in range(1, 4):
    cluster = np.zeros((1, 3))

    for i in range(m):
        for j in range(3):
            distancia[0, j] = np.linalg.norm(base[i, :] - neuronas[j, :])

        un = 0.5 * (1 - i / m)
        min_distancia = np.min(distancia)

        if distancia[0, 0] == min_distancia:
            neuronas[0, :] = neuronas[0, :] + un * (base[i, :] - neuronas[0, :])
            cluster[0, 0] += 1
            pertenencia[i] = 1
        elif distancia[0, 1] == min_distancia:
            neuronas[1, :] = neuronas[1, :] + un * (base[i, :] - neuronas[1, :])
            cluster[0, 1] += 1
            pertenencia[i] = 2
        elif distancia[0, 2] == min_distancia:
            neuronas[2, :] = neuronas[2, :] + un * (base[i, :] - neuronas[2, :])
            cluster[0, 2] += 1
            pertenencia[i] = 3
```

```python
# Print Neurons
neuronas_df = pd.DataFrame(neuronas, columns=BaseAumentada.columns)
neuronas_df
```

```python
# Print means
suma_promedio = np.sum(neuronas, axis=1) / 30
suma_promedio_df = pd.DataFrame({"Neurona": ["Neurona1", "Neurona2", "Neurona3"], "Suma.Promedio": suma_promedio})
suma_promedio_df
```

```python
# Print Clusters
cluster_df = pd.DataFrame({"Neurona": ["Neurona1", "Neurona2", "Neurona3"], "Cluster": cluster.flatten()})
cluster_df
```

```python
# Print Ownership
pertenencia_df = pd.DataFrame(pertenencia, columns=["Pertenencia"])
base_actualizada = pd.concat([pertenencia_df, BaseAumentada], axis=1)
base_actualizada
```
